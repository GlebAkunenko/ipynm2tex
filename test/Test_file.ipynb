# Plot section

There is a code that generate an image as output.



```python
import matplotlib.pyplot as plt

def f(x):
  return x * x

X = [i / 5 for i in range(0, 50)]
Y = [f(x) for x in X]

plt.plot(X, Y)
plt.show()
```
